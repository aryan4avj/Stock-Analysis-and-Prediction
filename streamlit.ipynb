```python
%%writefile app.py

import streamlit as st
import yfinance as yf
import pandas as pd

st.title("Stock Market Analysis")

#sidebar

# st.file_uploader("Upload CSV")

st.sidebar.title("Stocks")

st.sidebar.subheader("Stock Search")

# st.sidebar.selectbox("Example",["Facebook","Apple","Alphabet"])
stock_search = st.sidebar.multiselect('Stock Search', ["Facebook","Apple","Alphabet"])

start_date = st.sidebar.date_input('Start Date')
end_date = st.sidebar.date_input('End Date')

# st.write(start_date)

data = yf.download("AAPL", start=start_date, end=end_date)

data = data.drop["Volume","Close"]

st.line_chart(data)
```
